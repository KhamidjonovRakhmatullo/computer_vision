```python
# Import the downloader tool from the bing_image_downloader library
# We will use this downloader to search and download images from Bing

from bing_image_downloader import downloader
```

```python
# Download images from Bing using the search keyword: "Uzbek plov food"
# This will download images related to Uzbek plov / palov

downloader.download(
    "Uzbek plov food",                 # Search keyword for Bing images
    limit=5,                           # Number of images to download
    output_dir="/content/pilaf_images",# Folder path where images will be saved
    adult_filter_off=True,             # Allow downloader to search without adult filter restriction
    force_replace=False,               # Do not replace existing downloaded images
    timeout=60                         # Wait maximum 60 seconds for each image download
)
```

```python
# !rm -rf /content/pilaf_images
```

```python
# Import Google Drive tool from Google Colab
# This allows us to connect our Google Drive with Colab

from google.colab import drive

# Mount Google Drive inside Colab
# After running this, Google Drive files will be available in /content/drive

drive.mount('/content/drive')
```

```python
# Import required libraries
# os -> work with folders/files
# shutil -> move/copy/delete files

import os, shutil


# Source folder: images downloaded by bing-image-downloader
source_dir = "/content/pilaf_images/Uzbek plov food"

# Target folder: where we want to save images in Google Drive
target_dir = "/content/drive/MyDrive/uzbek_food_dataset/pilaf"


# Create target folder if it does not exist
# exist_ok=True means: do not show error if folder already exists
os.makedirs(target_dir, exist_ok=True)


# Loop through all files inside the source folder
for filename in os.listdir(source_dir):

    # Full path of the current image in source folder
    # ex: "/content/pilaf_images/Uzbek plov food/image_1.jpg"
    src_file = os.path.join(source_dir, filename)

    # Full path where the image will be moved in target folder
    # ex: "/content/drive/MyDrive/uzbek_food_dataset/pilaf/image_1.jpg"
    dst_file = os.path.join(target_dir, filename)

    # Move image from source folder to Google Drive folder
    shutil.move(src_file, dst_file)
```
